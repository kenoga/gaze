```python
# -*- coding: utf-8 -*-
import cv2
import os, sys
import glob
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
from utils import dataset_utils
```

```python
paths_dict = dataset_utils.get_image_paths(img_type="aligned_face")
fdf_dict = dataset_utils.get_face_direction_feature_dict()
```

```python
for pid in sorted(paths_dict.keys()):
    for path in paths_dict[pid][:5]:
        fname = os.path.basename(path)
        feature = fdf_dict[pid][fname]
        print(fname, feature)
        
        img = cv2.imread(path)
        for landmark in feature:
            x, y = [int(e) for e in landmark]
            cv2.circle(img, (x,y), 5, (255, 255, 255), 1)
        dataset_utils.imshow(img)
    break
```
