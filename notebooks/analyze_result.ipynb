```python
% matplotlib inline
import json
# import pandas
import matplotlib.pyplot as plt

def show_result(json_path):
    with open(json_path, 'r') as fr:
        data = json.load(fr)

    # Set values
    x = list(range(1, len(data['train']['loss'])+1))
    y0 = data['train']['loss']
    # y1 = data['train']['accuracy']
    y2 = data['test']['loss']
    y3 = data['test']['fscore']
    # y3 = data['test']['accuracy']
    # Set background color to white
    fig = plt.figure(figsize=(5,5),dpi=100)
    fig.patch.set_facecolor('white')

    # Plot lines
    plt.xlabel('epoch')
    plt.plot(x, y0, label='train_loss')
    # plt.plot(x, y1, label='train_accuracy')
    plt.plot(x, y2, label='test_loss')
    plt.plot(x, y3, label='test_fscore')
    plt.legend()

    # Visualize
    plt.show()
    plt.close()

    print('max f1 score: %f' % max(data['test']['fscore']))
```
