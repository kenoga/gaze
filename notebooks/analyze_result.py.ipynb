```python
% matplotlib inline
import json
import pandas
import matplotlib.pyplot as plt
```

## default

```python
with open('../train_eyecontact_detector/results/aligned_eyes_gray_equ_imgsize45-255_placesABCD_testids1-2-3_nonlockedrate1_batchsize128_learnrate0.001.json', 'r') as fr:
    data = json.load(fr)


# Set values
x = list(range(1, len(data['train']['loss'])+1))
y0 = data['train']['loss']
# y1 = data['train']['accuracy']
y2 = data['test']['loss']
y3 = data['test']['fscore']
# y3 = data['test']['accuracy']
# Set background color to white
fig = plt.figure(figsize=(5,5),dpi=100)
fig.patch.set_facecolor('white')

# Plot lines
plt.xlabel('epoch')
plt.plot(x, y0, label='train_loss')
# plt.plot(x, y1, label='train_accuracy')
plt.plot(x, y2, label='test_loss')
plt.plot(x, y3, label='test_fscore')
plt.legend()

# Visualize
plt.show()
plt.close()
```

## 学習データ数を変化させたときのf1スコアの変化

```python
# Set values
fpath = "../train_eyecontact_detector/results/f1_scores.txt"
df = pandas.read_csv(fpath, header=None)
fig = plt.figure(figsize=(5,2),dpi=100)
fig.patch.set_facecolor('white')
x = df[0]
y = df[1]
plt.xlabel('sample')
plt.ylabel('f1')
plt.plot(x, y, '--o', label='fscore')
plt.show()
plt.close()
```
