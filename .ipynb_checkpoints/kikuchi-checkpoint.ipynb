```python
import numpy as np
from urllib.request import urlopen
from io import BytesIO
from skimage import io, draw

with urlopen('http://i.ytimg.com/vi/oq20w95OlSY/hqdefault.jpg') as S:
  file = BytesIO(S.read())
img = io.imread(file)
io.imshow(img)
```

```python
H, W, _ = img.shape
hankei = 100
rr, cc = draw.circle(H//2, W//2, hankei)
img2 = np.copy(img)
img2[rr, cc, :] = (255, 255, 255)
io.imshow(img2)
```

```python
img3 = np.ones(img.shape, dtype=np.uint8) * 255
img3[rr, cc, :] = img[rr, cc, :]
io.imshow(img3)
```
