```python
for movie_file in movie_files:
    movie_id = movie_file.split(".")[0]
    src_sub_dir = os.path.join(src_dir, movie_id)
    tgt_sub_dir = os.path.join(tgt_dir, movie_id)
    if not os.path.exists(tgt_sub_dir):
        os.makedirs(tgt_sub_dir)
        
    json_path = os.path.join(json_dir, "%s.json" % movie_id)
    with open(json_path, "r") as fr:
        face_detection_dict = json.load(fr)
    result_path = os.path.join(result_dir, "%s.json" % movie_id)
    with open(result_path, "r") as fr:
        gaze_detection_dict = json.load(fr)
        
    src_img_paths = glob.glob(os.path.join(src_sub_dir, "*.jpg"))
    for i, src_img_path in enumerate(sorted(src_img_paths)):
        print("%d / %d" % (i, len(src_img_paths)))
        vis_img = visualize(src_img_path, face_detection_dict, gaze_detection_dict)
        img_name = os.path.basename(src_img_path)
        tgt_img_path = os.path.join(tgt_sub_dir, img_name)
        utils.imwrite(tgt_img_path, vis_img)
utils.notify("@nogaken visualization done!")
```
