```python
import os
import math
import cv2
import numpy as np
import time
```

```python
from IPython.display import display, Image

def imshow(image, format='.png'):
    decoded_bytes = cv2.imencode(format, image)[1].tobytes()
    display(Image(data=decoded_bytes))
```

```python
src_dir = './data/kobas-omni-eyecontact/image/0002'
fnames =  sorted(os.listdir(src_dir))
src_path = os.path.join(src_dir, fnames[0])
```

```python
size =  2880
img = cv2.imread(src_path)
print(img.shape)
img = cv2.resize(img, (size, size))
print(img.shape)
# imshow(img)
```

```python
from cam360.polar import PolarTransformer
polar_transformer = PolarTransformer(2880, 2880)
```

```python
import time
start = time.time()
polar_img = polar_transformer.transform(img, 0.5, 1, 0.75, 1)
need_time = time.time() - start
imshow(polar_img)
cv2.imwrite("./tmp/para_%04d.jpg" % size, polar_img)
cv2.imwrite("./tmp/omni_%04d.jpg" % size, img)
print(need_time)
```
