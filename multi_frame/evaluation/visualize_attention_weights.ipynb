```python
import os
import pickle
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cmap = sns.color_palette("Blues", 24)
fps=20
test_ids = [("05_01", "A"), ("05_01", "C"), ("05_01", "D"), ("05_02", "A"), ("05_02", "C"), ("05_02", "D")]
```

```python
for test_id in test_ids:
    session_id = "_".join(test_id)
    target_path = "./test_result/attention_weights/%s.pickle" % session_id
    weights = pickle.load(open(target_path))
    weights_tail = weights[:, 16:32]
    plt.figure()
    plt.subplots(figsize=(5,300))
    sns.heatmap(weights_tail, cmap=cmap, vmin=0, vmax=1)
    fpath = os.path.join("./figure/attention_weight", "%s.png" % session_id)
    plt.savefig(fpath)
```
