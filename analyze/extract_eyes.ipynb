```python
import os
import sys
import json

import cv2
```

```python
from IPython.display import display, Image
import cv2

def imshow(image, format='.png'):
    decoded_bytes = cv2.imencode(format, image)[1].tobytes()
    display(Image(data=decoded_bytes))
```

```python
para_root_dir = "./data/kobas-omni-eyecontact/transformed_image"
face_root_dir = "./data/kobas-omni-eyecontact/face_image"
face_sub_dirs = sorted([d for d in os.listdir(face_root_dir) if d[0] != '.'])
face_dirs = [os.path.join(face_root_dir, sub) for sub in face_sub_dirs]
fd_result_fname = "face_position_and_landmarks.json"
dst_root_dir = "./data/kobas-omni-eyecontact/both_eyes"

if not os.path.exists(dst_root_dir):
    os.mkdir(dst_root_dir)
```

```python
def extract_both_eyes(img, lms):
    eyes_top = lms[37][1]
    eyes_bot = lms[41][1]
    eyes_left = lms[36][0]
    eyes_right = lms[45][0]
    margin = 20
    eyes_img = img[eyes_top-margin:eyes_bot+margin, eyes_left-margin:eyes_right+margin]
#     cv2.imwrite(os.path.join(eyes_dir, fname), eyes_img)
    return eyes_img
```

```python
for face_sub_dir in face_sub_dirs:
    face_dir = os.path.join(face_root_dir, face_sub_dir)
    fd_result_path = os.path.join(face_dir, fd_result_fname)
    with open(fd_result_path, "r") as fr:
        fd_result = json.load(fr)
        for fname in fd_result.keys():
                if not fd_result[fname]["detected"]:
                    continue
                landmarks = fd_result[fname]["lanmarks"]
                
                # 12のpathを変えた
                fname_s = fname.split(".")[0].split("_")
                if int(fname_s[0]) == 12 and fname_s[2] in {"B", "C"}:
                    fname_s[0] = "0017"
                    para_img_path = os.path.join(para_root_dir, "0017", "_".join(fname_s)+".jpg")
                else:
                    para_img_path = os.path.join(para_root_dir, face_sub_dir, fname)
                    
                para_img = cv2.imread(para_img_path)
                if para_img is None:
                    print("Can not read %s." % para_img_path)
                    continue
                eye_img = extract_both_eyes(para_img, landmarks)


                dst_dir = os.path.join(dst_root_dir, face_sub_dir)
                if not os.path.exists(dst_dir):
                    os.mkdir(dst_dir)
                
                dst = os.path.join(dst_dir, fname)
                cv2.imwrite(dst, eye_img)
```
