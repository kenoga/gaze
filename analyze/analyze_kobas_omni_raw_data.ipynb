```python
import os
from collections import defaultdict
data_root_dir = "./data/kobas-omni-eyecontact/image/"
data_sub_dirs = os.listdir(data_root_dir)
data_dirs = sorted([os.path.join(data_root_dir, data_sub_dir) for data_sub_dir in data_sub_dirs if data_sub_dir[0] != '.'])

id2cnt = defaultdict(int)
glasses2cnt = defaultdict(int)
place2cnt = defaultdict(int)
target2cnt = defaultdict(int)

for data_dir in data_dirs:
    fnames = [name.split('.')[0].split('_') for name in os.listdir(data_dir)]
    for fname in fnames:
        id2cnt[fname[0]] += 1
        glasses2cnt[fname[1]] += 1
        place2cnt[fname[2]] += 1
        target2cnt[fname[3]] += 1

print("sum")
s = sum([v for k, v in id2cnt.items()])
print(s)

print("people_id statistics")
for key, value in sorted(id2cnt.items()):
    print(key, value)
    
print("glasses statistics")
for key, value in sorted(glasses2cnt.items()):
    print(key, value)

print("place statistics")
for key, value in sorted(place2cnt.items()):
    print(key, value)
    
print("target statistics")
for key, value in sorted(target2cnt.items()):
    print(key, value)
    
    
```
