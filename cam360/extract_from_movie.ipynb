```python
src_root_dir = '../data/kobas-omni-eyecontact/movie/'
dst_root_dir = '../data/kobas-omni-eyecontact/image/'
src_dirs = [os.path.join(src_root_dir, child) for child in sorted(os.listdir(src_root_dir))]
dst_dirs = [os.path.join(dst_root_dir, child) for child in sorted(os.listdir(src_root_dir))]
if not os.path.exists(dst_root_dir):
    os.mkdir(dst_root_dir)
for dst_dir in dst_dirs:
    if not os.path.exists(dst_dir):
        os.mkdir(dst_dir)
```

```python
# -*- coding: utf-8 -*-
import os
import numpy as np
from io import BytesIO
import codecs
import cv2

def extract_under_half(frame):
    height = frame.shape[0]
    return frame[0:height//2, :]
    
def extract_left(frame):
    width = frame.shape[1]
    return frame[:, 0:width//2]
def do_nothing(frame):
    return frame

def mov2imgs(in_path, out_path, func=do_nothing, interval=2):
    # func: 動画から得られた画像に対してかける処理
    cap = cv2.VideoCapture(in_path)
    if not cap.isOpened():
        raise Exception("動画の読み込みに失敗しました。")
        
    in_movie_name = os.path.basename(in_path)
    frame_count = 0
    
    while True:
        # interval+1フレームおきに保存する
        for _ in range(0,interval):
            _, _ = cap.read()
        ret, frame = cap.read()
        frame_count += 1
        if not ret:
            break
        
        frame = func(frame)
        img_name = "_".join([in_movie_name.split('.')[0],  "%08d" % frame_count]) + ".jpg"
        print(in_movie_name + "->" + img_name)
        
        cv2.imwrite(os.path.join(out_path, img_name), frame)
    cap.release()
```

```python
for src_dir, dst_dir in zip(src_dirs[5:], dst_dirs[5:]):
    fnames = os.listdir(src_dir)
    for fname in fnames:
        mov2imgs(os.path.join(src_dir, fname), dst_dir, interval=2)
```
